```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t, r):
    c = 1 
    d = 5
    y1, y2, y3 = r
    d_y1_dt = -c * y1 * y2
    d_y2_dt = c * y1 * y2 - d * y2
    d_y3_dt = d * y2
    return d_y1_dt, d_y2_dt, d_y3_dt
```

```python
t = np.linspace(0, 1, 1000)
sol = integrate.solve_ivp(f, (0,1), (95, 5, 0), t_eval = t)
y_1, y_2, y_3 = sol.y
plt.title("Model Kermack’a-McKendrick’a")
plt.ylabel("f(t) - % populacji")
plt.xlabel("t")
plt.plot(t, y_1, label="$y_1(t)$ - Osoby zdrowe")
plt.plot(t, y_2, label="$y_2(t)$ - Osoby zainfekowane")
plt.plot(t, y_3, label="$y_3(t)$ - Osoby ozdrowiałe")
plt.legend()
plt.legend(bbox_to_anchor=(0.53, -0.15))
```

```python
def universal_plot(title, x, y, xlabel, ylabel, llabel=None, scale=False, planet=None):
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.plot(x, y, label=llabel)
    if planet:
        plt.scatter(planet[0], planet[1], label="Okrążana planeta", c="black") 
    if llabel:
        plt.legend()
        plt.legend(bbox_to_anchor=(0.53, -0.15))
    if scale:
        plt.axis("equal")
    plt.show()
```

```python
def f(t, vec):
    x, dx, y, dy = vec
    r = np.sqrt(x**2 + y**2)
    return dx, -x/(r**3), dy, -y/(r**3)
```

```python
max_t = 6.4
t = np.linspace(0, max_t, 10**6)
e = 0.9
energy = []
momentum = []
sol = integrate.solve_ivp(f, (0, max_t), (1-e, 0, 0, ((1+e) / (1-e))**(1/2)), t_eval = t, rtol=1e-8, atol=1e-8)
x, dx, y, dy = sol.y
```

```python
universal_plot(f"Wykres x(t) dla e = {e}", sol.t, x, "t", "x(t)")
```

```python
universal_plot(f"Wykres y(t) dla e = {e}", sol.t, y, "t", "y(t)")
```

```python
universal_plot(f"Wykres x(t) i y(t) dla mimośrodu = {e} w problemie dwóch ciał", x, y, "x(t)", "y(t)", None , True, None)
```

```python
energy = []
momentum = []
for i in range(len(x)):
    r = np.sqrt(x[i]**2 + y[i]**2)
    energy.append((dx[i]**2 + dy[i]**2)/2 - 1/r)
    momentum.append(x[i]*dy[i] - y[i]*dx[i])
```

```python
universal_plot("Wykres całkowitej energii układu w zależności od t dla e = 0.9", np.linspace(0,max_t,len(energy)),energy, "t", "Energia")
```

```python
universal_plot("Wykres całkowitego pędu układu w zależności od t dla e = 0.9", np.linspace(0,max_t,len(momentum)),momentum, "t", "Pęd")
```
