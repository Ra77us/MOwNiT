```python
from sympy import *
from sympy.abc import x
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def plot(f, xmin, xmax, const=0, p=1000):
    x_vals = np.linspace(xmin, xmax, p)
    y_vals = [f.subs(x, v) for v in x_vals]
    y_0  = [const for i in x_vals]
    plt.plot(x_vals, y_vals, label="Badana funkcja")
    plt.plot(x_vals,y_0, label=f"y={const}", c="black")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def calculate(start, g, iterations):
    res = start
    res_v = []
    for i in range(iterations):
        res = g.subs(x,res).evalf()
        print(f"After {i+1} iterations x = ", res.evalf())
        res_v.append(res)
    return res_v
```

```python
def get_r_vector(res_v, expected):
    r_v = []
    epsilon_v = [abs(r - expected) for r in res_v]
    for i in range(2, len(epsilon_v)):
        r_v.append(math.log(epsilon_v[i-1] / epsilon_v[i]) / math.log(epsilon_v[i-2] / epsilon_v[i-1]))
        print(f"for k = {i} r = ", r_v[-1])
```

```python
def newton(f, der_f, start, iterations):
    prev_x = start
    for i in range(iterations):
        next_x = prev_x - f.subs(x, prev_x) / der_f.subs(x, prev_x)
        print(f"x_{i}:", next_x.evalf())
        print(f"|x_{i} - x_{i-1}|", abs(prev_x - next_x).evalf())
        prev_x = next_x
```

```python
prev_x = np.array([0.5,0.5])
for i in range(5):
    x1 = prev_x[0]
    x2 = prev_x[1]
    next_x = prev_x - np.linalg.inv(np.matrix([[2*x1, 2*x2], [2*x1, - 1]])) @ np.array([x1**2 + x2**2-1, x1**2-x2]).T
    prev_x  = np.array([next_x[0,0],next_x[0,1]])
    print(f"After {i+1} iteration(s)",prev_x)
```

```python
f = x**3 - 2*x - 5
plot(f,-5,5)
der_f = 3*x**2 -2
start = 3
newton(f, der_f, start, 4)
```

```python
f = np.e**-x -x
der_f = -1 - np.e**-x
plot(f,-5,20)
start = -1
epsilon = 10**(-5)
newton(f, der_f, start, 4)
```
